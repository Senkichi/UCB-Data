```python
import pandas as pd
from statistics import mean
```

```python
gamer_data = pd.read_csv('purchase_data.csv')
gamer_data.head()
```

```python
player_count = gamer_data['SN'].nunique()
unique_items = gamer_data['Item ID'].nunique()
avg_item_price = gamer_data['Price'].mean()
total_purchases = len(gamer_data['Price'])
total_revenue = gamer_data['Price'].sum()
player_count
```

```python
#Gender Demographics
gendered_gamers = gamer_data.drop_duplicates(subset = 'SN')
gendered_gamer_pct = (gendered_gamers['Gender'].value_counts()/len(gendered_gamers))*100
gendered_gamers_count = gendered_gamers['Gender'].value_counts()

```

```python
#Purchasing analysis
gender_group = gamer_data.groupby('Gender')
gender_purchase_count = gender_group['SN'].count()
gender_avg_purchase = gender_group['Price'].mean()
gender_total_purchase = gender_group['Price'].sum()
```

```python
#Age Demographics
age_bin = [0, 10, 14, 18, 22, 26, 30, 34, 38, 42, 46]
age_bin_names = ['<10', '10-14', '14-18', '18-22', '22-26', '26-30', '30-34', '34-38', '38-42', '42-46']
gamer_data['Why four?'] = pd.cut(gamer_data['Age'], age_bin, labels = age_bin_names)
binned_gamers = gamer_data.groupby('Why four?').agg({'Price': ['count', 'mean', 'sum']})
binned_gamers
```

```python
#Top Spenders
top_spenders = gamer_data.groupby('SN').sum()
top_spenders = top_spenders.sort_values('Price', ascending = False)
top_sns = top_spenders.reset_index()
top_sns = [top_sns['SN'][0:5]]
top_sns2 = ['Lisosia93', 'Idastidru52', 'Chamjask73', 'Iral74', 'Iskadarya95']
spendy_records = gamer_data.loc[gamer_data['SN'].isin(top_sns2)]
spendy_records
spendy_group = spendy_records.groupby('SN').agg({'Price': ['count', 'mean', 'sum']})
spendy_group
```

```python
#Top Items
top_items = gamer_data.groupby('Item Name').count()
top_items = top_items.sort_values('Price', ascending = False)
top_items = top_items.reset_index()
top_items
top_five_items = top_items['Item Name'][0:5]
top_five_items =  ['Final Critic', 'Oathbreaker, Last Hope of the Breaking Storm', 'Persuasion', 'Nirvana', 'Extraction, Quickblade Of Trembling Hands']
item_records = gamer_data.loc[gamer_data['Item Name'].isin(top_five_items)]
item_records = item_records.groupby(['Item ID', 'Item Name']).agg({'Price': ['count', 'mean', 'sum']})
item_records
```
