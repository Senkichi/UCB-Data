```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
from splinter import Browser
import re
import selenium as se
from requests_html import HTMLSession
import zipfile
import urllib
import os
```

```python
url = "https://s3.amazonaws.com/tripdata/index.html"
```

```python
driver = se.webdriver.Chrome(executable_path='C:/Users/samue/chromedriver.exe')
driver.get(url)
```

```python
soup=BeautifulSoup(driver.page_source)
soup
```

```python
zip_links = []
```

```python
for link in soup.find_all('a'):
    zip_links.append(link.get('href',None))
zip_links
```

```python
bike_df = pd.DataFrame()
```

```python
zip_links[0]
```

```python
date = re.search(r'/(\d+)',zip_links[0])
```

```python
date.group(1)
```

```python
zip_ref = zipfile.ZipFile(zip_links[0], 'r')
```

```python
for link in zip_links:
    
    #get date
    date = re.search(r'/(\d+)',link).group(1)
    
    #directory sorting
    zip_name = f'./JC-{date}citibike-tripdata.csv.zip'
    destination = f'./csv_files'
    csv_name = f'./JC-{date}citibike-tripdata.csv'
    
    urllib.request.urlretrieve(link,zip_name)
    print('--------------')
    print(f'{zip_name} saved')
    
    
    #unzip
    zip_ref = zipfile.ZipFile(zip_name, 'r')
    zip_list = zip_ref.namelist()    
    zip_ref.extract(zip_list[0])
    zip_ref.close()
    os.rename(zip_list[0], csv_name)
    print(f'{zip_name} unzipped')

    ###append to dataframe and add month starting column
    file_df = pd.read_csv(csv_name)
    file_df['date'] = f'{date}'

    #force headers to match first df
    if len(bike_df)>0:
        file_df.columns = bike_df.columns.values.tolist()
    else:
        file_df

    bike_df = bike_df.append(file_df)


    #remove zip file and temp csv
    os.remove(zip_name)
    os.remove(csv_name)

    print(f'{zip_name} complete')   
```

```python
bike_df.to_csv(f'{destination}/CitiBike_Master.csv')
```
